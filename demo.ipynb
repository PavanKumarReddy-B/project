## Loading data from url

## creating csv files using url

```python
import json
import csv

from urllib.request import urlopen

with urlopen("https://o136z8hk40.execute-api.us-east-1.amazonaws.com/dev/get-list-of-conferences") as response:
    source = response.read()
    data = json.loads(source)
    fname="display_paid.csv"
    fname1="display_free.csv"

    with open(fname, "w") as file:
        csv_file= csv.writer(file)
        csv_file.writerow(["emailId","city","twitter_handle","keywordSupport","country","imageURL","venue","searchTerms","confName","state","long","confEndDate","conference_id","lat","user_id","confUrl","confStartDate","entryType","confRegUrl"])
        for item in data["paid"]:
            csv_file.writerow([item['emailId'],item['city'],item['twitter_handle'],item['keywordSupport'],item['country'],item['imageURL'],item['venue'],item['searchTerms'],item['confName'],item['state'],item['long'],item['confEndDate'],item['conference_id'],item['lat'],item['user_id'],item['confUrl'],item['confStartDate'],item['entryType'],item['confRegUrl']])

    with open(fname1, "w") as file:
        csv_file= csv.writer(file)
        csv_file.writerow(["emailId","city","twitter_handle","keywordSupport","country","imageURL","venue","searchTerms","confName","state","long","confEndDate","conference_id","lat","user_id","confUrl","confStartDate","entryType","confRegUrl"])
        for item in data["free"]:
            csv_file.writerow([item['emailId'],item['city'],item['twitter_handle'],item['keywordSupport'],item['country'],item['imageURL'],item['venue'],item['searchTerms'],item['confName'],item['state'],item['long'],item['confEndDate'],item['conference_id'],item['lat'],item['user_id'],item['confUrl'],item['confStartDate'],item['entryType'],item['confRegUrl']])
```

## Importing pandas

```python
import pandas as pd
```

## Reading the csv files

```python
display_paid = pd.read_csv('display_paid.csv', encoding = "ISO-8859-1")
display_free = pd.read_csv('display_free.csv', encoding = "ISO-8859-1")
```

## Concatenating two csv files into one file

```python
data_update=pd.concat((display_paid, display_free))
```
